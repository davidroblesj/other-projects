**Problema 1.48**


Una cadena que pertenece al Lenguaje $D$ es aquella que tiene el mismo numero de ocurrencias del substring 01 y 10. Esto sucede cuando tenemos los mismos cambios del substring 0 al substring 1 asi como del substring 1 al substring 0.

Ejemplo:

En la cadena: 101, tenemos un cambio de 1 a 0 y de 0 a 1, tenemos los mismos cambios, por lo tanto pertenece al lenguaje.

En la cadena: 00110010, tenemos un cambio de 0 a 0, no se cuenta, de 0 a 1, de 1 a 1, no se cuenta, de 1 a 0, de 0 a 0, no se cuenta, de 0 a 1 y de 1 a 0.

Una cadena que cumpla con lo anterior siempre terminara empezará y terminará con el mismo simbolo. Entonces podemos usar inducción matematica para demostrar que la propiedad anterior descrita es cierta.

Donde $n$ es igual a la longitud de la cadena.

**Caso base $n = 1$:**

Tomamos las cadenas que terminan con el mismo simbolo con el que empiezan que son 0 y 1 solamente y se cumple ya que hay 0 ocurrencias del substring 01 y 10.

Para $n = 2$:

Tenemos las cadenas 00 y 11, se cumple ya que hay 0 ocurrencias del substring 01 y 10.

**Hipótesis de inducción: Las cadenas de longitud $n$ que terminan con el mismo simbolo con el que empiezan pertenecen al lenguaje $D$.**

*Por demostrar: Se cumple lo anterior para longitud $n + 1$*

Sea $u$ una cadena de longitud $n + 1$ del lenguaje $D$

Sea $u'$ la cadena de longitud $n$ igual a $u$ menos el ultimo caracter de $u$ 

Tenemos dos casos:

- Caso 1: $u'$ pertenece a D

Entonces, por hipótesis de inducción $u'$ empieza con el mismo elemento que con el que termina y pertenece a D. 


Supongamos que $u'$ empieza y termina con 0 o 1 y $u$ termina con 1 o 0 respectivamente, entonces $u$ es igual a agregarle 1 o 0 a $u'$ pero esto le agregaria una ocurrencia del substring 01 o 10, por lo tanto $u$ ya no seria

parte del lenguaje D, esto es una contradiccion, por lo tanto la unica posibilidad es que $u$ termine con el mismo elemento que $u'$, es decir termine y empiece con el mismo elemento

- Caso 2: $u'$ no pertenece a D

Por hipótesis de inducción $u'$ no empieza con el mismo elemento que termina.

Si $u'$ termina con 0 o 1 y $u$ termina igual con 0 o 1 respectivamente, $u$ no pertenecería a $D$ ya que $u'$ no pertenece a D y agregar el mismo elemento no agregaria la ocurrencia del substring 01 o 10 que falta. Por lo tanto es una contradiccion y el unico caso es que $u$ termine con el mismo elemento con el que inicia.

Por lo tanto $u$ termina con el mismo elemento con el que inicia.

Ahora si tenemos el alfabeto $\sigma = {0,1}$ la expresion regular para las cadenas en $\sigma$ que termina con el mismo elemento con el que inicia es:

0(0|1)∗0|1(0|1)∗1 

Por lo tanto existe un automata finito equivalente a la expresion regular, que acepta las cadenas de D, por lo tanto D es regular

```python
import itertools
def automata_a_regular(automata):
    '''
    Funcion automata_a_regular
    
    Convierte un automata finito a su expresion regular equivalente
    
    Recibe:
    
    automata: af.AFN o af.AFD
    
    Regresa:
    
    expr_regl: string
    
    '''
    q_1q_2, q_1q, qq, qq_2 = '','','',''
    programa = list(automata.programa)
    estado_inicial = automata.estado_inicial
    estados_de_aceptacion = list(automata.estados_de_aceptacion)
    #si el estado inicial es un estado de aceptacion o tiene transiciones hacia el
    #creamos un estado inicial con una transicion epsilon hacia el estado inicial
    #original
    if estado_inicial in estados_de_aceptacion or any(i.siguiente == estado_inicial for i in programa):
        programa.append(af.Transicion("inicial","$",estado_inicial))
        estado_inicial = 'inicial'   
    #si hay mas de una estado de aceptacion, creamos un solo estado de aceptacion
    #y unimos con transicion epsilon al nuevo estado de aceptacion a todos los
    #estados de aceptacion antiguos
    if len(estados_de_aceptacion)>1:
        estados_de_aceptacion = ['aceptacion']
        for i in automata.estados_de_aceptacion:
            programa.append(af.Transicion(i,"$",estados_de_aceptacion[0]))
    elif any(i.estado == estados_de_aceptacion[0] for i in programa):
        estados_de_aceptacion = ['aceptacion']
        for i in automata.estados_de_aceptacion:
            programa.append(af.Transicion(i,"$",estados_de_aceptacion[0]))
    estados = list(automata.estados)
    if estado_inicial == 'inicial':
        estados.append('inicial')
    if 'aceptacion' in estados_de_aceptacion:
        estados.append('aceptacion')
    for estado in estados:#iteramos en los estados a eliminar
        if estado not in estados_de_aceptacion and estado != estado_inicial:
            #solo eliminamos un estado sino es de aceptacion o inicial
            duplet = itertools.product(estados,estados)
            #generamos todas las dupletas de los estados
            for par in duplet:
                if par[0]!=estado and par[1]!=estado:
                    #estado es q, par[0] es q_1 y par[1] es q_2
                    for transicion in programa:
                        if transicion.estado == par[0] and transicion.siguiente == par[1]:
                            q_1q_2 = transicion.simbolo
                        if transicion.estado == par[0] and transicion.siguiente == estado:
                            q_1q = transicion.simbolo
                        if transicion.estado == estado and transicion.siguiente == estado:
                            qq = transicion.simbolo
                        if transicion.estado == estado and transicion.siguiente == par[1]:
                            qq_2 = transicion.simbolo
                    if q_1q=='' or qq_2 =='':
                        #si no se encontro una ocurrencia de q_1q o qq_2 saltamos la iteracion
                        q_1q_2, q_1q, qq, qq_2 = '','','',''
                        continue                 
                    for transicion in programa[:]:
                        if transicion.estado == par[0] and transicion.siguiente == par[1]:
                            programa.remove(transicion)
                    if q_1q_2 == '' or qq == '':
                        #agregamos las transicion necesaria
                        if q_1q_2 == '' and qq == '':
                            programa.append(af.Transicion(par[0],(q_1q + qq_2) ,par[1]))
                        elif qq == '':
                            programa.append(af.Transicion(par[0],(q_1q_2 + '|' +  qq_2) ,par[1]))
                        elif q_1q_2 == '':
                            programa.append(af.Transicion(par[0],(q_1q + '(' + qq +')' + '*' + qq_2) ,par[1])) 
                    else:
                        programa.append(af.Transicion(par[0],(q_1q_2 + '|' + q_1q + '(' + qq +')' + '*' + qq_2) ,par[1]))
                    q_1q_2, q_1q, qq, qq_2 = '','','',''
            for t in programa[:]:
                #eliminamos el estado
                if (t.estado == estado) or (t.siguiente == estado):
                    programa.remove(t)
    expr_regl = programa[0].simbolo.replace('$','')
    return expr_regl
```

```python
def gramatica_a_pila(grammar, t, nt, inicial):
    '''
    Funcion gramatica_a_pila
    
    Recibe:
    grammar: diccionario (producciones)
    t: conjunto (terminales)
    nt: conjunto (no terminales)
    inicial: string (produccion inicial)
    
    Regresa:
    
    P: lista de tuplas (automata de pila)
    
    '''
    P = [(0,'','',1,'$'),
         (1,'','',2,inicial)]
    estados = 2 #contador de estados
    for produccion in grammar:
        produce = grammar.get(produccion)
        for cadena in produce:
            for caracter in reversed(cadena):
                if caracter == cadena[-1]:
                    #si estamos al final de nuestra cadena
                    estados = estados + 1
                    P.append((2, '',produccion,estados,caracter))
                elif caracter == cadena[0]:
                    #al principio de nuestra cadena
                    P.append((estados, '','',2,caracter))
                else:
                    #en medio de nuestra cadena
                    estados = estados + 1
                    P.append((estados-1, '','',estados,caracter))
                if len(cadena) == 1:
                    #caso especial cuando el inicio y el final es igual
                    P.remove((2, '',produccion,estados,caracter))
                    estados = estados - 1
                    P.append((2, '',produccion,2,caracter))
    for terminal in t:
        #agregamos las transiciones de terminales
        estado = (2, terminal,terminal,2,'')
        P.append(estado)
    P.append((2, '','$',estados+1,''))          
    return P
```
